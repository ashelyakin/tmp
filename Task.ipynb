```python
import math
import numpy
import scipy.stats
import matplotlib.pyplot
%matplotlib inline
```

```python
# параметр k для распределения хи-квадрат
k = 4.0
chi2 = scipy.stats.chi2(k)
```

```python
# выборка объёмом 1000
rvs = chi2.rvs(1000)
```

```python
# построим гистограмму выборки и график плотности распределения хи-квадрат
matplotlib.pyplot.hist(rvs, normed=True)
x = numpy.linspace(0, 17.5, 100)
matplotlib.pyplot.plot(x, chi2.pdf(x))
matplotlib.pyplot.xlabel('$x$')
matplotlib.pyplot.ylabel('$f(x)/n$')
```

```python
# зададим значения n
n0 = 5
n1 = 10
n2 = 50
```

```python
# составим массив выборочных средних из выборок объёма n0
avs0 = []
for i in xrange(1000):
    rvs0 = chi2.rvs(n0)
    sum = 0.0
    for j in xrange(n0):
        sum += rvs0[j]
    avs0.append(sum / n0)
```

```python
# для распределения хи-квадрат матожидание равно k, а дисперсия - 2k
# по центральной предельной теореме, параметрами аппроксимирующего нормального распределения являются:
# матожидание - k
# среднеквадратическое отклонение - дисперсия, делённая на объём выборок - 2k / n0
norm0 = scipy.stats.norm(k, math.sqrt(2 * k / n0))
```

```python
matplotlib.pyplot.hist(avs0, normed=True)
matplotlib.pyplot.plot(x, norm0.pdf(x))
matplotlib.pyplot.xlabel('$x$')
matplotlib.pyplot.ylabel('$f(x)/n$')
```

```python
# аналогично для n1
avs1 = []
for i in xrange(1000):
    rvs1 = chi2.rvs(n1)
    sum = 0.0
    for j in xrange(n1):
        sum += rvs1[j]
    avs1.append(sum / n1)
```

```python
norm1 = scipy.stats.norm(k, math.sqrt(2 * k / n1))
```

```python
matplotlib.pyplot.hist(avs1, normed=True)
matplotlib.pyplot.plot(x, norm1.pdf(x))
matplotlib.pyplot.xlabel('$x$')
matplotlib.pyplot.ylabel('$f(x)/n$')
```

```python
# аналогично для n2
avs2 = []
for i in xrange(1000):
    rvs2 = chi2.rvs(n2)
    sum = 0.0
    for j in xrange(n2):
        sum += rvs2[j]
    avs2.append(sum / n2)
```

```python
norm2 = scipy.stats.norm(k, math.sqrt(2 * k / n2))
```

```python
matplotlib.pyplot.hist(avs2, normed=True)
matplotlib.pyplot.plot(x, norm2.pdf(x))
matplotlib.pyplot.xlabel('$x$')
matplotlib.pyplot.ylabel('$f(x)/n$')
```

Вывод: нормальным распределением достаточно хорошо приближается распределение выборочных средних. При этом с ростом объёма выборок возрастает и точность аппроксимации.
